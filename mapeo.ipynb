```python
from pylab import *
%run  "mis_integradores.ipynb"
%matplotlib inline
```

```python
def osc(y,t,w02=1.):
    return array([y[1],-w02*sin(y[0])+0.1*sin(t)])
```

```python
figure(figsize=(10,10))
for i in range(3):
    y0,u0 = pi*(2*rand()-1),0 #4*(2*rand()-1)
    t, sol = RK4(osc,[y0,u0],[0,500.])
    plot(sol[:,0],sol[:,1])
zoom = pi
xlim([-zoom,zoom]), ylim([-zoom,zoom])
axes().set_aspect('equal'), grid(True)
show()
```

```python
figure(figsize=(10,10))
y0,u0 = 1.2,0
t, sol = RK4(osc,[y0,u0],[0,10000.])
plot(sol[:,0],sol[:,1])
y0,u0 =1.21,0
t, sol1 = RK4(osc,[y0,u0],[0,10000.])
plot(sol1[:,0],sol1[:,1])
zoom = pi
xlim([-zoom,zoom]), ylim([-zoom,zoom])
axes().set_aspect('equal'), grid(True)
show()
```
