# EULER PARA ECUACIONES DIFERENCIALES DE PRIMER ORDEN

```python
def euler(funcion, p_ini, t_final, h=0.01):
    tiempos = arange(p_ini[0],t_final+h,h)
    y = zeros(len(tiempos))
    y[0] = p_ini[1]
    for i in range(len(tiempos)-1):
        y[i+1] = y[i] + funcion(y[i],tiempos[i])*h
    solucion = [tiempos, y]
    return solucion
```

Esta es la ecuaci√≥n diferencial a resolver:
$$f'(x) = f(x)$$

```python
def mi_ec_diff(y,t):
    return y 
```

```python
sol = euler(mi_ec_diff,[0,1],3.,.001)
plot(sol[0],sol[1],sol[0],exp(sol[0]))
```

Esta es la ecuaci√≥n diferencial a resolver:
$$f'(x) = -xf(x)$$

```python
def otra_func(y,t):
    return -y*t
```

```python
sol = euler(otra_func,[0,1],3.,.001)
plot(sol[0],sol[1],sol[0],exp(-(sol[0]**2)/2))
```

```python
def erandi(f,x):
    return (1./(sin(x)+2))*(((2.*f)/sqrt(3))**2-1)
```

```python
hs = [0.5,0.1,0.01,0.001,1e-3]
for h in hs:
    sol = euler(erandi,[1,1],3.,h)
    plot(sol[0],sol[1])
show()
```

# EULER PARA ECUACIONES DIFERENCIALES DE MAYOR ORDEN

```python
def euler(funcion, p_ini, t_final, h=0.01):
    tiempos = arange(p_ini[0],t_final+h,h)
    y = [0]*len(tiempos)
    y[0] = p_ini[1]
    for i in range(len(tiempos)-1):
        y[i+1] = y[i] + funcion(y[i],tiempos[i])*h
    solucion = [tiempos, y]
    return solucion
```
